```python
import networkx as nx
import csv
import matplotlib.pyplot as plt
```

```python
class ladefensebuilding:
    def __init__(self,name,function,height,delivery,zone,node,longitude,latitude):
        self.building = building
        self.function = function
        self.height = height
        self.delivery = delivery
        self.zone = zone
        self.node = node
        self.longitude = longitude
        self.latitude = latitude
       
        
```

```python
with open("ladefensebuilding.csv", encoding='mac_roman') as csvfile:
    reader = csv.reader(csvfile)
    rows = 0
    for r in reader:
        rows = rows + 1
    print(rows)
```

```python
ladefensebuildings=[]
with open("ladefensebuilding.csv", encoding='mac_roman') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        building = row['Building']
        function = row['function']
        height = row['height']
        delivery = row['Delivery']
        zone = row['zone']
        node = row['node']
        longitude = row['longitude']
        latitude = row['latitude']

        ladefensebuildings.append(ladefensebuilding(building,function,height,delivery,zone,node,longitude,latitude))
```

```python
ladefensebuildinggraph=nx.Graph()
Pathgraph = nx.Graph()
Stronggraph = nx.DiGraph()

#ladefensebuildinggraph.pop = {}  # size
```

```python
ladefensebuildingposition = {}  # location
i=0
while i<rows-1:
    ladefensebuildingposition[i]=(float(ladefensebuildings[i].longitude),float(ladefensebuildings[i].latitude))
    i=i+1
print(ladefensebuildingposition)
```

```python
for P in ladefensebuildings:
    ladefensebuildinggraph.add_node(P)
```

```python
for P1 in ladefensebuildings:
    for P2 in ladefensebuildings:
        if not P1 == P2:
            #if int(P1.height)>100 and int(P2.height)>100:
               # ladefensebuildinggraph.add_edge(P1,P2)
                
               # Pathlist=list(nx.shortest_path(ladefensebuildinggraph,P1,P2))
               # print(Pathlist)
               # for a in Pathlist:
                   # Pathgraph.add_edge(P1, P2)
            if P1.zone == P2.zone:
                ladefensebuildinggraph.add_edge(P1,P2) 
            
            if P1.function == P2.function:
                    ladefensebuildinggraph.add_edge(P1,P2)
                    #Stronglist=list(nx.strongly_connected_components(Stronggraph))
                    #print(Stronglist)
                    #for c in Stronglist: 
                    #    Stronggraph = nx.cycle_graph(c, create_using=nx.DiGraph())            
```

```python
pos = nx.spring_layout(ladefensebuildinggraph)
node_labels = {}
for node in ladefensebuildinggraph.nodes:
    node_labels[node] = str(node.building)
```

```python
options={'node_color':'red','node_size':15,'width':0.5,'edge_color':'black'}
options3 = {'node_color':'green', 'edge_color':'green', 'node_size':10, 'width':1}
options2 = {'node_color':'blue', 'edge_color':'orange', 'node_size':20, 'width':1}
plt.figure(1, figsize = (20,20))
nx.draw(Stronggraph, **options2)
nx.draw(Pathgraph, **options3)
nx.draw(ladefensebuildinggraph,ladefensebuildingposition)
nx.draw_networkx_labels(ladefensebuildinggraph, pos, labels=node_labels)

```
